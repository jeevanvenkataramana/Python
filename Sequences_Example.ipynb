### Finding 8 directions by indices

```python
def right_direction(row_len, col_len, row_index, col_index):
        print("right")
        co_ord=[(row_index,x) for x in range(col_index,col_len)]
        print(co_ord)    


def left_direction(row_len, col_len, row_index, col_index):
    print("left")
    co_ord=[(row_index,x) for x in range(col_index,-1,-1)]
    print(co_ord)
    
def top_direction(row_len, col_len, row_index, col_index):
    print("top")
    co_ord=[(x,col_index) for x in range(row_index,-1,-1)]
    print(co_ord)

def down_direction(row_len, col_len, row_index, col_index):
    print("down")
    co_ord=[(x,col_index) for x in range(row_index,row_len)]
    print(co_ord)


def left_slant_down(row_len, col_len, row_index, col_index):
    print("left slant down")
    x_ord=[x for x in range(row_index,row_len)]
    y_ord=[y for y in range(col_index,col_len)]
    min_len=len(y_ord)
    if len(x_ord)<len(y_ord):
        min_len=len(x_ord)
    co_ord=[(x_ord[x],y_ord[x]) for x in range(min_len)]
    print(co_ord)

    
def left_slant_up(row_len, col_len, row_index, col_index):
    print("left slant up")
    x_ord=[x for x in range(row_index,-1,-1)]
    y_ord=[y for y in range(col_index,-1,-1)]
    min_len=len(y_ord)
    if len(x_ord)<len(y_ord):
        min_len=len(x_ord)
    co_ord=[(x_ord[x],y_ord[x]) for x in range(min_len)]
    print(co_ord)
    
def right_slant_up(row_len, col_len, row_index, col_index):
    print("Right slant up")
    x_ord=[x for x in range(row_index,-1,-1)]
    y_ord=[y for y in range(col_index,col_len)]
    min_len=len(y_ord)
    if len(x_ord)<len(y_ord):
        min_len=len(x_ord)
    co_ord=[(x_ord[x],y_ord[x]) for x in range(min_len)]
    print(co_ord)

def right_slant_down(row_len, col_len, row_index, col_index):
    print("Right slant down")
    x_ord=[x for x in range(row_index,row_len)]
    y_ord=[y for y in range(col_index,-1,-1)]
    min_len=len(y_ord)
    if len(x_ord)<len(y_ord):
        min_len=len(x_ord)
    co_ord=[(x_ord[x],y_ord[x]) for x in range(min_len)]
    print(co_ord)
    
    

right_direction(5,6,1,5)
left_direction(5,6,1,5)
top_direction(5,6,1,5)
down_direction(5,6,1,5)
left_slant_down(5,6,1,5)
left_slant_up(5,6,1,5)
right_slant_up(5,6,1,5)
right_slant_down(5,6,1,5)

print("\n\n\n")
```
