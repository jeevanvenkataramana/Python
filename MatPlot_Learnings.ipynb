## MatPlot Practice

### Import library

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Sample plot

```python
x=np.arange(0,10,2)
y=x+2
print(x)
print(y)
plt.plot(x,y)                                              # plot(x_axes, y_axes, label)
plt.xlabel('X axes')
plt.ylabel('Y axes')
plt.title('Sample')
plt.show()
```

```python
### sample plot 2 mark points
```

```python
x = np.linspace(0, 2, 10)
mark = {'marker':"x"}
plt.figure(1)
plt.plot(x, x, label='linear',**mark)
mark['marker']='.'
plt.figure(2)
plt.plot(x, x**2, label='quadratic',**mark)
mark['marker']='*'
plt.figure(3)
plt.plot(x, x**3, label='cubic',**mark)
plt.legend()                                                  # displays sublabels in graph
plt.show()
```

```python
x = np.arange(0, 10, 0.2)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

### several plots in same graph

```python
x=np.arange(1,5)
plt.plot(x,x,x,x*2,x,x*3)
plt.show()
```

### Create subplots

```python
x = np.arange(0, 10, 0.2)
y = np.sin(x)
z=np.cos(x)

fig,(ax1,ax2)=plt.subplots(1,2)
plt.suptitle('Sin and Cos Function')        # main title of two sub graphs   subplots(no of rows, no of columns)
ax1.plot(x,y)
ax2.plot(x,z)
```

### Mark Points

```python
x = np.arange(0, 10, 0.2)
y = np.sin(x)
fig,ax=plt.subplots(1,1)
ax.plot(x,y,**{'marker': 'x'})
plt.suptitle("sin function")
```

### define scale of x axis and Y axis   -- plt.axis([x_min,x_max,y_min,y_max])

```python
x=[2,4,8,16]
y=[3,9,27,81]
plt.plot(x,y,'b')
plt.axis([0,20,0,100])
plt.show()
```

### Format styles

```python
x=[2,4,8,16]
y=[3,9,27,81]
plt.plot(x,y,'r')
plt.axis([0,20,0,100])
plt.show()
```

```python
x=[2,4,8,16]
y=[3,9,27,81]
plt.plot(x,y,'ro')
plt.axis([0,20,0,100])
plt.show()
```

```python
x=[2,4,8,16]
y=[3,9,27,81]
plt.plot(x,y,'g')
plt.axis([0,20,0,100])
plt.show()
```

```python
x=[1,2,3,4,5]
x=np.array(x)
plt.plot(x, x,'r--', x, x**2, "bs", x, x**3, 'g^')   # red dashes, Blue squares and green triangles

plt.show()
```
